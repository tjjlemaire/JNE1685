# Figure 4 - Effective variables

Create the figure panels describing the model's effective variables dependencies on US frequency, US amplitude and sonophore radius.

### Matplotlib parameters

```python
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
matplotlib.rcParams['font.family'] = 'arial'
```

### Imports

```python
import os
import matplotlib.pyplot as plt
from PySONIC.plt import plotEffectiveVariables
from PySONIC.utils import logger
from PySONIC.neurons import getPointNeuron
from utils import saveFigsAsPDF
```

### Plot parameters

```python
figindex = 4
fs = 12
lw = 2
ps = 15
figs = {}
```

### Simulation parameters

```python
pneuron = getPointNeuron('RS')
a = 32e-9  # m
Fdrive = 500e3  # Hz
Adrive = 50e3  # Pa
```

## Panel A: dependence on acoustic amplitude

```python
fig = plotEffectiveVariables(pneuron, a=a, Fdrive=Fdrive, cmap='Oranges', zscale='log')
figs['a'] = fig
```

## Panel B: dependence on US frequency

```python
fig = plotEffectiveVariables(pneuron, a=a, Adrive=Adrive, cmap='Greens', zscale='log')
figs['b'] = fig
```

## Panel C: dependence on sonophore radius

```python
fig = plotEffectiveVariables(pneuron, Fdrive=Fdrive, Adrive=Adrive, cmap='Blues', zscale='log')
figs['c'] = fig
```

### Save figure panels

Save figure panels as **pdf** in the *figs* sub-folder:

```python
saveFigsAsPDF(figs, figindex)
```
